```python
import os
import datetime
import pause
import time
import re
import pandas as pd
from selenium.webdriver import FirefoxProfile, Firefox
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import Select, WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from fake_useragent import UserAgent
```

```python
site = 'https://matricula.ufabc.edu.br' # target website

base = pd.read_excel("Base.xlsx", header = 0, decimal = ",") # reads the excel file with the user data and subjects to be enrolled
print(base.columns, "\n")
print(f"{base['Usuario'][0]} \n {base['Senha'][0]} \n\n {base['Materias'].values}\n")

user_id = base['Usuario'][0] # gets the user id
user_passw = base['Senha'][0] # gets the user password
materias = base['Materias'].values # gets the subjects to be enrolled

ua = UserAgent() # instantiates the fake user agent
user_agent = ua.random # gets a random user agent
opts = Options() # sets some settings regarding the way the webscrapping will run

opts.add_argument(f'--user-agent={user_agent}') # sets the user agent to a random one to avoid detection from captcha

opts.set_preference("profile",os.path.expanduser(r'~\AppData\Roaming\Mozilla\Firefox\Profiles\cqfxjsa7.default-release')) # sets the profile to be used by the webdriver
opts.set_preference("dom.webdriver.enabled", False) 
opts.set_preference('useAutomationExtension', False)

browser = Firefox(options=opts) # instantiates the selenium Firefox webdriver with the specified options 

wait = WebDriverWait(browser, 600) # sets the webdriver to wait for 10 minutes before raising an error

now = datetime.datetime.now() # gets the current date and time
today = datetime.date.today().strftime('%d/%m/%Y') # gets the current date
then = datetime.datetime.strptime(f'{today} 12:00:00', '%d/%m/%Y %H:%M:%S') # sets the time to start the webscraping
diff = then - now # calculates the difference between the current time and the time to start the webscraping

if then > now: # if the time to start the webscraping is greater than the current time, the program will wait until the time to start
    print(f'\nToday: {today}')
    print(f'Current Time: {now.time()}')
    print(f'Time to start webscraping: {then.time()}')
    print(f'Time to wait until execution: {diff.total_seconds() + 40}')

    print('\nWaiting...')
    time.sleep(diff.total_seconds() + 40)

print('\nStarting...\n')

errors = 0
while True:
    if errors == 5: # if the webdriver fails to access the website 5 times, the program will stop
        print('Too many errors, exiting...')
        browser.close()
        break
    try:
        browser.get(site) # makes the webdriver access the given website
        break
    except:
        errors += 1
        browser.refresh() # refreshes the webdriver to try again

browser.maximize_window() # maximizes the browser window to avoid problems with the website layout

errors = 0
while True:
    if errors == 5: # if the webdriver fails to access the website 5 times, the program will stop
        print('Too many errors, exiting...')
        browser.close()
        break
    try:
        browser.find_element(By.ID, "uid").send_keys(user_id) # finds the element to input the user id and sends the user id
        browser.find_element(By.ID, "senha").send_keys(user_passw) # finds the element to input the user password and sends the user password

        # ActionChains(browser).key_down(Keys.TAB).key_up(Keys.TAB).key_down(Keys.ENTER).key_up(Keys.ENTER).perform()
        # time.sleep(10)

        # browser.find_elements(By.CLASS_NAME, "btn-success")[0].click()

        wait.until(EC.element_to_be_clickable((By.ID, "aplicarfiltros"))) # waits until the button to apply the filters is clickable

        for i in materias: # selects the subjects to be enrolled
            browser.find_element(By.XPATH, f"//span[contains(text(), '{i}')]").find_element(By.XPATH, '..').find_element(By.XPATH, '..').find_element(By.TAG_NAME, "input").click()

        send_btt = browser.find_element(By.ID, "enviar")

        ActionChains(browser).move_to_element(send_btt).perform() # moves the cursor to the button to send the request

    except:
        errors += 1
        browser.refresh() # refreshes the webdriver to try again
```
