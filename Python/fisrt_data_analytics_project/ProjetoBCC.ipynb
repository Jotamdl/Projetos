##Pesquisa na base de dados

```python
#Ferramenta para pesquisar informações sobre sobre espécies epecíficas dentro da base de dados:
x = input("Deseja fazer pesquisas baseado no *nome da espécie* ou em sua *linha* respectiva na tabela? ")
es = ['espécie','especie','ESPECIE','ESPÉCIE','nome da espécie','Nome da espécie','nome da especie','Nome da especie']
li = ['linha','LINHA','Linha']
if x in es:
  x = ""
  while x != "0":
    x = input("Digite o nome da espécie que você procura ou digite 0 para encerrar o programa: ")
    filtro = df[df["Espécie (Simplificado)"] == x]
    if x == "0":
      print("Fim da execução\n")
    elif filtro.empty == True:
      print("Espécie não consta na lista\n")
    else:
      print("\nESPÉCIE:"+x+"\n", filtro.iloc[0] ,"\n")
else:
  x = ""
  while x != "00":
    x = input("Digite a linha da espécie que você procura ou digite 00 para encerrar o programa: ")
    if x == "00":
      print("Fim da execução\n")
    else:
      x = int(x)
      if x > 3285:
        print("Linha não consta na lista\n")
      else:
        y = df.loc[x,"Espécie (Simplificado)"]
        print("\nESPÉCIE:"+y+"\n", df.loc[x] ,"\n")
```

```python
#Ferramenta para observar o nivel de proteção de espécies específicas dentro da base de dados:
x = input("Digite o nome de alguma espécie ou digite 0 para encerrar o programa: ")
if x == '0':
  print('\nPrograma encerrado')
while x != '0':
  x1 = df[df["Espécie (Simplificado)"] == x]
  if x1.empty == True:
    print("Espécie não consta na lista\n")
  else:
    y = x1.iloc[0,12]
    if y == 1:
      print('A espécie',x,'possui proteção nula\n\n')
    elif y == 2 or y == 3:
      print('A espécie',x,'possui proteção fraca\n\n')
    else:
      print('A espécie',x,'possui proteção forte\n\n')
  x = input("Digite o nome de alguma espécie ou digite 0 para encerrar o programa: ")
  if x == '0':
    print('\nPrograma encerrado')
```

```python
#Ferramenta para comparar estados quanto ao seu número de casos de epécies ameaçadas:
x = input('Digite algum estado ou 0 caso deseje encerrar o programa: ')
if x == "0":
  print('\nPrograma encerrado')
while x != '0':
  v1 = c12[c12.index==x]
  if v1.empty == True:
    print("Estado não consta na lista\n")
  else:
    v1 = v1.loc[x,'Estados de Ocorrência']
    y = input('Digite outro estado: ')
    v2 = c12[c12.index==y]
    if v2.empty == True:
      print("Estado não consta na lista\nReiniciando programa\n\n")
    else:
      v2 = v2.loc[y,'Estados de Ocorrência']
      if v1 > v2:
        print('\n',x,'possui mais espécies ameaçadas do que', y,'\n')
      else:
        print('\n',y,'possui mais espécies ameaçadas do que', x,'\n')
  x = input('\nDigite algum estado ou 0 caso deseje encerrar o programa: ')
  if x == "0":
   print('\nPrograma encerrado')
```

```python
#Ferramenta para comparar tipos de ameaça quanto ao número de epécies ameaçadas pela mesma:
x = input('Digite alguma ameaça ou 0 caso deseje encerrar o programa: ')
if x == "0":
  print('\nPrograma encerrado')
while x != '0':
  v1 = c6[c6.index==x]
  if v1.empty == True:
    print("Tipo de ameaça não consta na lista\n")
  else:
    v1 = v1.loc[x,"Principais Ameacas"]
    y = input('Digite outra ameaça: ')
    v2 = c6[c6.index==y]
    if v2.empty == True:
      print("Outro tipo de ameaça não consta na lista\nReiniciando programa\n\n")
    else:
      v2 = v2.loc[y,"Principais Ameacas"]
      if v1 > v2:
        print('\n',x,'possui maior impacto sobre as espécies do que', y,'\n')
      else:
        print('\n',y,'possui maior impacto sobre as espécies do que', x,'\n')
  x = input('\nDigite alguma ameaça ou 0 caso deseje encerrar o programa: ')
  if x == "0":
   print('\nPrograma encerrado')
```
